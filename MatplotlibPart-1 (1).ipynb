```python
import matplotlib.pyplot as plt
import IPython
print(IPython.sys_info())
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)

y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method for creating plot
plt.plot(x,y,'r')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')

#plt.show()
```

```python
plt.subplot(1,2,1) # (1,2,1) means 1 row,2 col,plot num
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
# oo method for creating plot
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #(left,bottom,height,wide)
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])#inset axes
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows = 3,ncols = 3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')
#plt.tight_layout()
```

** plot Apperence **

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'orange',linewidth = 10)# defult linewidh 1  
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'orange',lw = 10,alpha = 0.5)# defult linewidh 1  
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'orange',lw = 3,linestyle = 'steps')# defult linewidh 1 and (linestyle = ':' or linestyle = '--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'orange',lw = 3,ls = '-',marker = 'o')# as len(x) = 11 so we can get 11 dot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'orange',lw = 0.5,ls = '-',marker = '2',markersize = 10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = '#FF8C00') RGB Hex code
ax.plot(x,y,color = 'purple',lw = 1,ls = '-',marker = 'o',markersize = 20,
       markerfacecolor = 'yellow',markeredgewidth = 3,markeredgecolor = 'green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'orange',lw = 2,ls = '--')
ax.set_xlim([0,1])#x limit lower and upper bound
ax.set_ylim([0,2])#y limit lower and upper bound
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
